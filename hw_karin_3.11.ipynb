```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


X = np.array([
    [2, 15, 40, 0],
    [5, 16, 45, 1],
    [3, 16, 40, 0],
    [10, 18, 50, 5],
    [7, 17, 45, 3],
    [1, 14, 35, 0],
    [8, 16, 45, 4],
    [4, 15, 40, 1],
    [6, 15, 42, 2],
    [12, 19, 55, 8]
])
Y = np.array([15, 25, 18, 45, 35, 12, 38, 22, 30, 60])


model = LinearRegression()
model.fit(X, Y)


print(f"Intercept (0): {model.intercept_:.2f}")
print(f"Coefficients (1..4): {model.coef_}")


Y_pred = model.predict(X)


Y_mean = np.mean(Y)
sse = np.sum((Y - Y_pred) ** 2)
sst = np.sum((Y - Y_mean) ** 2)
r2_manual = 1 - sse / sst

print(f"SSE: {sse}")
print(f"SST: {sst}")
print(f"R (manual): {r2_manual}")


print(f"R (model.score): {model.score(X, Y)}")
print(f"MSE: {mean_squared_error(Y, Y_pred)}")
print(f"R (r2_score): {r2_score(Y, Y_pred)}")

new_worker = np.array([[6, 16, 43, 2]])  
predicted_salary = model.predict(new_worker)
print(f"Expected salary: {predicted_salary[0]:.2f} ")
```

```python
Most influential: Managerial experience (highest coefficient, 2.57)
Next: Professional experience (1.97)
Less influential: Weekly working hours (0.20)
Almost no influence: Education (0.12)

#The salary is mainly affected by the employees overall experience and managerial background.
#In other words, someone with many years of experience- especially in management- will earn significantly more,
#even if they work fewer hours or have slightly fewer years of education.
#Education and working hours add a bit, but they dont make a big difference overall.
```
