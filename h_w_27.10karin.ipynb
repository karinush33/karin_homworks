```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
x = np.arange(-5,10)
#x + y = 12
y1 = 12-x
#x -y = 4
y2 = x-4

y1
y2
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x,y1, label = "y = 12-x")
plt.plot(x,y2, label = "y = x -4")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.annotate(f"(8, 4)",
             (8,4),
             textcoords="offset points",
             xytext=(0,20),
             ha='center',
             fontsize=10,
             color='blue')
plt.scatter(8,4, color = "green", zorder = 5)
plt.show()
print ("single solution")
```

```python
x = np.arange(-5,10)
#5x+ 3y= 12
y1 = 4.3-1.6*x
#4x-2y=6
y2 = 2*x-3

y1
y2
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x,y1, label = "y =4.3 - 1.6 * x")
plt.plot(x,y2, label = "y = 2 * x - 3")
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.annotate(f"(2, 1)",
             (2,1),
             textcoords="offset points",
             xytext=(0,20),
             ha='center',
             fontsize=10,
             color='blue')
plt.scatter(2,1, color = "green", zorder = 5)
plt.show()
print ("single solution")
```

```python
x = np.arange(-5,10)

# 2x + 4y = 8
y1 = 2 - 0.5*x

# x + 2y = 4
y2 = 2 - 0.5*x  

plt.axhline(0, color='black', linewidth=0.8, linestyle='--') 
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  
plt.plot(x, y1, label='2x + 4y = 8')
plt.plot(x, y2, label='x + 2y = 4', linestyle='--') 

plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()


plt.scatter([0,2,4], [2,1,0], color='green', zorder=5)

plt.text(0,3.5, "Infinite solutions", color='blue')

plt.show()
print ("Infinite solution")
```
