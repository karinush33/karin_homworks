```python
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
# 1. Linear Model â€“ Taxi Fare vs Distance ðŸš•

x = np.arange(0, 11)
y = 8 + 5 * x

plt.title('Taxi Fare vs Distance ') 
plt.plot(x, y)         
plt.legend(["price"])      
plt.xlim(0, 10)            
plt.ylim(0, 60)    
plt.savefig('taxi_fare_line.jpg')
plt.show()
```

```python
# 2. Visual Analysis with tips.csv ðŸ½ï¸

df = pd.read_csv("tips.csv")

#2.1. Scatter: price_per_person (x) vs tip (y)
#Create a scatter plot of price_per_person (x-axis) vs tip (y-axis)
#Add title and axis labels

plt.scatter(df['price_per_person'], df['tip'])

plt.title('price_per_person vs tip') 
plt.xlabel('price per person')
plt.ylabel('tip')
plt.show()

#By looking at the graph- Is there a correlation?
'''
 NO-There does not appear to be a clear relationship between price per person and tip, 
 as the points are scattered without a distinct trend.
```

```python
# 2.2. Bar: maximum total_bill per day

df = pd.read_csv("tips.csv")
# For each day, find the max total_bill
# Plot a bar chart of these maxima

max_bills = df.groupby('day')['total_bill'].max()
plt.bar(max_bills.index, max_bills.values)
```

```python
# 2.3. Histogram: tips

df = pd.read_csv("tips.csv")
# Plot a histogram of tip

plt.hist(df['tip'], bins='auto', color='purple')
plt.show()

# Question: By looking at the graph- what is the range of the most common tip?
'''
Most tips fall within the range of 2 to 3 dollars, making this the most common tip amount.
```

```python
#2.4. Histogram: tip percentage

df = pd.read_csv("tips.csv")

#Create a new column tip_perc = 100 * tip / total_bill

df['tip_perc'] = 100 * df['tip'] / df['total_bill']
#Plot a histogram of tip_perc
plt.hist(df['tip_perc'], bins=30, color='orange')
plt.show()
#Question: By looking at the graph- what is the range of the most tip percentage?
'''
Most tip percentage fall within the range of 10 to 20 dollars, making this the most common tip amount.
'''
```

```python
# 2.5. Subplots: combine all four visuals
#Create a 2Ã—2 grid using fig, axes = plt.subplots(2, 2, figsize=(10, 7))
df = pd.read_csv("tips.csv")
fig, graphs = plt.subplots(nrows=2, ncols=2, figsize=(10,7))

# Recreate the four plots in their own axes objects: (1) scatter, (2) bar of max per day, (3) tip histogram, (4) tip% histogram
# Make sure each subplot has a title and labeled axes
display(graphs.shape)

graphs[0,0].scatter(df['price_per_person'], df['tip'])
graphs[0,0].set_xlabel('price_per_person')
graphs[0,0].set_ylabel('tip')
graphs[0,0].set_title('price_per_person vs tip') 

max_bills = df.groupby('day')['total_bill'].max()
graphs[0,1].bar(max_bills.index, max_bills.values)
graphs[0,1].set_xlabel('day')
graphs[0,1].set_ylabel('max_total_bill')
graphs[0,1].set_title("Maximum Total Bill per Day")

graphs[1,0].hist(df['tip'], bins=30, color='purple')
graphs[1,0].set_title("Distribution of Tips")
graphs[1,0].set_xlabel("Tip amount")
graphs[1,0].set_ylabel("Frequency")

df['tip_perc'] = 100 * df['tip'] / df['total_bill']
graphs[1,1].hist(df['tip_perc'], bins='auto', color='orange')
graphs[1,1].set_title("Distribution of Tip Percentage")
graphs[1,1].set_xlabel("Tip %")
graphs[1,1].set_ylabel("Frequency")

# Adjust layout and save as tips_overview.png
plt.savefig('tips_overview.png')

plt.tight_layout()
plt.show()
```
